computer to solve MIS with a unit-disk radius for connectivity larger than kingâ€™s.

We want to take into account the following things: $\Omega$ needs to be large in order to effectively mix the ground state, $\Delta$ needs to become bigger than $\Omega$ such that the Ground State of the Hamilonian coincides with the MIS, $\Delta$ needs to increase adiabatically so that the ground state follows the evolution of the hamiltion by the adiabatic theorem. Finally, Delta is chosen as negative in the beginning to ensure that the ground state is |00000...0>. The requirement "connectivity larger than king's" just means $$R_b/a > 2$$. The next qualitatevely different case (new "distance" added) occurs at ~2.11 (geometric mean of 2 and $\sqrt 5$).

In order to implement e.g. 2.11, we choose $\Omega$ = 15.7 MHZ, $\Delta_{Max} = (c_6 / a *  2.11))^{1/6} = 15$ MHz. The driving frequency is chosen to increase rapidly, as a rectangular function, to maximise the mixing interaction mediated by the operator preceded by $\Omega(t)$.

- Determine if the parameters above can fit a machine like Aquila. If so, generate problem instances on Bloqade that could be run on it.
Extra points for finding if Rb/a=3 is realizable experimentally on Aquila, or what would be the best performant largest value.

We came to the conclusions that, given Aquila's capabilities, a ratio of 3 is not implementable on Aquila. The main limit is the fact that "a" is lower bounded by 4 $\mu m$, and therefore the Rydberg Radius is bounded by 12  $\mu m$. This however translates to a $\Delta$ which is too small to extract the solution of the MIS as the ground state, namely because the system will not enter the state.
